Metrics for the 1 vs all classification

```python
path = Path('../dataset/1_vs_all/')
```

```python
tfms = get_transforms(max_rotate=30, max_zoom=1.3, max_lighting=0.4, max_warp=0,
                     p_affine=1., p_lighting=1.)
```

```python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
                                 ds_tfms=tfms, size=224, bs=32, num_workers=3).normalize()
```

```python
learner = cnn_learner(data, models.densenet121, pretrained=False, metrics=[accuracy,error_rate])
```

```python
learner.load('Dense-121-stage-1-MP-WG');
```

```python
learner.to_fp16();
```

```python
interp = ClassificationInterpretation.from_learner(learner)
```

```python
ytrue = interp.y_true.cpu()
ypred = interp.pred_class.cpu()
probas = interp.probs.cpu()
```

```python
sp_weights = np.array(len(ytrue)/(2*np.bincount(ytrue)),dtype=np.float16)
```

```python
from fastai.callbacks.hooks import *
```

```python
def hooked_backward(cat=y):
    with hook_output(m[0]) as hook_a:
        with hook_output(m[0], grad=True) as hook_g:
            preds = m(xb)
            preds[0,int(cat)].backward()
    return hook_a, hook_g
```

```python
def show_heatmap(hm,tp='act'):
    fig, ax = plt.subplots(figsize=(8,8))
    xb_im.show(ax)
    ax.imshow(hm, alpha=0.6, extent=(0,224,224,0),
             interpolation='bilinear', cmap='jet',vmin=hm.min()*(1-0.2))
    ax.set_title('{}'.format(data.valid_ds.y[idx]))
    if tp == 'act':
        plt.savefig('act_map_{}.png'.format(data.valid_ds.y[idx]))
    elif tp == 'grad':
        plt.savefig('grad_map_{}.png'.format(data.valid_ds.y[idx]))
```
